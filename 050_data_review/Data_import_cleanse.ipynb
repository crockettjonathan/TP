```python
#import required libaries
import pandas as pd
from pathlib import Path
```

```python
#Read in CSV and review for nulls

# 1a. Get the directory of the notebook (in Jupyter)
current_dir = Path.cwd()
# 1b. Get the directory where the current script is located (non jupyter)
#current_dir = Path(__file__).resolve().parent

# 2. Define path to the sibling folder and source data
csv_path = current_dir.parent / "010_source_data" / "tp_reviews.csv"

tp_reviews=pd.read_csv(csv_path)
#replace spaces in the column names with an underscore for best practice naming conventions in DB
tp_reviews.columns = tp_reviews.columns.str.replace(' ', '_')
tp_reviews.count()
```

```python
#create list with Review IDs
Rid=tp_reviews["Review_Id"]
Rid
```

```python
#Identify duplicate reviews
tp_reviews[Rid.isin(Rid[Rid.duplicated()])].sort_values("Review_Id")
```

```python
#Review single instance
filter1="ccdbbec8-5496-4a1f-99dd-a81770f7a206"
tp_reviews[tp_reviews["Review_Id"]==filter1]
```

```python
#import libary for DB
import sqlite3

#define connection to DB
db_path = current_dir.parent / "100_prod" / "trust.db"
conn=sqlite3.connect(db_path)
```

```python
#create table in DB
tp_reviews.to_sql('sql_tp_reviews',conn,if_exists='replace',index=False)
```

```python
#Review Duplicates from the DB
pd.read_sql_query('SELECT * FROM sql_tp_reviews where Review_Id in (select Review_Id from sql_tp_reviews group by 1 having count(*)>1) order by 1', conn)
```

```python
#Remove duplicates and create deduplicated data within the database
tp_reviews_dedup=tp_reviews.drop_duplicates(subset=['Review_Id'])
#tp_reviews_dedup.count()
tp_reviews_dedup.to_sql('sql_tp_reviews_dedup',conn,if_exists='replace',index=False)
```
